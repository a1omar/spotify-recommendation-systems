## Normalization

First convert int types that will be included in normalization to float types

```python
data = data_prepared
```

```python
# turning int types to float types for normalization
data['key'] = data['key'].astype(float)
data['mode'] = data['mode'].astype(float)
```

Creating a new dataframe for features that will be normalized

```python
col_names = ["valence",
             "acousticness",
             "danceability",
             "energy",
             "instrumentalness",
             "key",
             "liveness",
             "loudness",
             "mode",
             "speechiness",
             "tempo"]
dataf = data[col_names]
```

## One-hot encoding

Performing one hot encoding

```python
year = pd.get_dummies(data['year'])
popularity = pd.get_dummies(data['popularity'])
```

```python
data_new = pd.concat([dataf,year,popularity,data['id']],axis = 1)
```

# Vectorization
- Term Frequency - Inverse Document Frequency (tf-idf)

When performing tf-idf for genres we need to make sure there are no null values present since tf-idf vectorizor cannot handle null values

Checking for null values in the data

```python
total_null = data['genre_list'].isnull().sum()

print(f"The total amount of null values for genres is {total_null}")
```

```python
data['genre_list'] = data['genre_list'].fillna("").apply(list)
```

```python
data['genre_list'].isnull().sum()
```

Perform tf-idf on genres

```python
# initalizing tf-idf
tf_idf = TfidfVectorizer()
```

```python
# creating matrix
tfidf_matrix =  tf_idf.fit_transform(data['genre_list'].apply(lambda x: " ".join(x)))
```

```python
# printing dataframe
genres = pd.DataFrame(tfidf_matrix.toarray())
```

```python
# naming each column feature
genres.columns = ['genre' + "_" + i for i in tf_idf.get_feature_names()]
```

```python
# reseting index
genres.reset_index(drop = True, inplace=True)
```
